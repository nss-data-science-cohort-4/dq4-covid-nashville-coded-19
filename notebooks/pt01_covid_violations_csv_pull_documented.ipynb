### Take the CSV of code violations downloaded from the hubNashville site, limit it to COVID-19 violations, and 

```python
import pandas as pd
```

```python
# import the service requests csv from hubNashville
service_requests = pd.read_csv('../data/hubNashville_311_service_requests.csv')
service_requests.head()
```

```python
# take a look at the unique values within the 'Request Type' field
service_requests['Request Type'].unique()
```

```python
# create a covid19 dataset with only reported violations that are in the COVID-19 request and subrequest types
covid19 = service_requests[service_requests['Request Type'] == 'COVID-19']
covid19 = service_requests[service_requests['Subrequest Type'] == 'COVID-19 Violations']
covid19 = covid19.reset_index(drop = True)
covid19.head()
```

```python
# Take a look at the shape of the file
covid19.shape
```

```python
# Take a look at the types of each variable
covid19.info()
```

```python
# Check out the unique values in selected fields to see if we should keep them or not
print(covid19['System of Origin ID'].unique())
print(covid19['Request Origin'].unique())
print(covid19['Contact Type'].unique())
print(covid19['Closed When Created'].unique())
print(covid19['Preferred Language'].unique())
print(covid19['Contact Type'].unique())
```

```python
# Figure out how many of the violations have a parent request and how many were closed when created
print(covid19['Parent Request'].value_counts())
print(covid19['Closed When Created'].value_counts())
```

```python
# Quick check to see if they are the same request - they are not
covid19[covid19['Parent Request'] == 339141.0]
```

We removed the one report that was `True` for `closed when created`, as well as the one violation with a `parent request` id, to ensure that the data were consistent.

```python
covid19 = covid19[covid19['Closed When Created'] == False]
```

```python
covid19 = covid19[covid19['Parent Request'].isna() == True]
```

```python
# Debug check
covid19.head()
```

```python
# Get all column names
covid19.columns
```

```python
# Get rid of unnecessary columns or columns with only NaN values
covid19 = covid19.drop(columns=['Status','Request Type','Subrequest Type','Additional Subrequest Type',
                       'Date / Time Closed', 'Request Origin', 'System of Origin ID', 'Preferred Language', 
                       'State Issue', 'Parent Request','Closed When Created','Council District'])
```

```python
# Debug check
covid19.head()
```

```python
# Get all of the new columns
covid19.columns
```

```python
# Rename the columns
covid19 = covid19.rename(columns={'Request #':'request_no','Date / Time Opened': 'datetime_opened',
                                  'Contact Type': 'contact_type', 'Address': 'address','City': 'city','ZIP': 'zip','Latitude': 'lat',
                                  'Longitude': 'long','Mapped Location': 'coord'})
```

```python
# Final check to make sure everything is still where we think it is
covid19.shape
```

```python
# Write the cleaned-up COVID-only dataset to a new csv
covid19.to_csv('../data/covid_violations.csv', index = False)
```
