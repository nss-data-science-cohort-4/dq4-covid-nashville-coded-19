```python
# import statements
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# display settings
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
violations_with_business_info = pd.read_csv('../data/violations_with_business_info.csv')
violations_with_business_info
```

```python
violations_with_business_info = violations_with_business_info.fillna(0)
violations_with_business_info
```

```python
violations_by_type = violations_with_business_info.sum(axis = 0).to_frame()
violations_by_type
```

```python
violations_by_type = violations_by_type.drop(['Unnamed: 0', 'request_no', 'datetime_opened', 'zip', 'lat', 'long'], axis = 0)
violations_by_type
```

```python
violations_by_type = violations_by_type.reset_index()
violations_by_type
```

```python
violations_by_type.columns = ['business_type', 'count_of_businesses']
violations_by_type
```

```python
violations_by_type = violations_by_type.sort_values(by = 'count_of_businesses', ascending = False).reset_index(drop = True)
```

```python
violations_by_type
```

```python
violations_by_type.columns
```

```python
violations_by_type = violations_by_type.drop([0,1], axis = 0)
```

```python
violations_by_type
```

```python
violations_by_type.info()
```

```python
violations_by_type['count_of_businesses'] = violations_by_type['count_of_businesses'].astype('int')
```

```python
violations_by_type.describe()
```

```python
# Take a look to see if there's a correlation between USEDMEM_PCT and USEDMEM_TOT
plt.figure(figsize=(10,6))
# make barplot
sns.barplot(x='business_type', y = 'count_of_businesses', data = violations_by_type[violations_by_type['count_of_businesses'] >= 143.75])
# set labels
plt.xlabel('Business Type', size=14)
plt.ylabel('Number of Reported Violations', size=14)
plt.title('Business Types with the Most Frequently-Reported COVID-19 Violations', size=18)
plt.xticks(rotation = 290)
plt.tight_layout()
plt.savefig('../visualizations/top_quartile_business_types_with_violations.png', dpi=150)
```
